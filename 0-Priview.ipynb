- 원 핫 인코딩

### 레이블 인코딩: Category Feature를 코드형 숫자값으로 변환

```python
from sklearn.preprocessing import LabelEncoder

items = ["Tv", "냉장고", "전자렌지", "컴퓨터", "선풍기", "선풍기", "믹서", "믹서"]

encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
```

```python
print(labels)
print(encoder.classes_)
print(encoder.inverse_transform([0,1,4,5,3,3,2,2]))
```

## 원핫 인코딩

- 새로운 feature를 추가해 고유 값에 해당하는 컬럼에만 1을 표시하고 나머지는 0으로 표시

```python
from sklearn.preprocessing import OneHotEncoder

items = ["Tv", "냉장고", "전자렌지", "컴퓨터", "선풍기", "선풍기", "믹서", "믹서"]
```

```python
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)

labes = labels.reshape(-1, 1)

onehot = OneHotEncoder()
onehot.fit(labels)
oh= onehot.transform(labels)

oh.toarray()
```

### 3) Feature Scaling

- 표준화: (Xi - mean(X)) / sd(X)
- 정규화: (Xi - mean(X)) / max(X) - min(x)

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

## StandardScaler

```python
from sklearn.datasets import load_iris

iris = load_iris()
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
print("feature들의 평균\n", df.mean())
print("------------------------------------------")
print("feature들의 분산\n", df.var())
```

```python
scaler=StandardScaler()
scaler.fit(df)
iris_scaled=scaler.transform(df)
iris_scaled
```

```python
df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_name)
```

## Scaling을 할 때 주의할 점

```python
train = np.arange(0,11).reshape(-1,1)
test=np.arange(0,6).reshape(-1,1)

print(train)
print(test)
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
train_scaled=scaler.transform(train)

print("원본: ", train.flatten())
print("스케일링: ", train_scaled.flatten())

print("-----------------------------------------------")

test_scaled = scaler.transform(test)

print("원본: ", train.flatten())
print("스케일링: ", train_scaled.flatten())
```
